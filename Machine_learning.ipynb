Linear regression algorithm

- Linear regression is a statisical method used to model the relationship  between a dependent(target) and an independent variable.
- The goal is to fit a line that best represents the data. The basic idea is to find a line that minimises the distnce between the predicted values and the actual values.
- Two types of linear regression:
1. Simple Linear Regression : Involves one independent variable and one dependent variable.
2. Multiple Linear Regresssion : Involves just one independent variable.

Assumptions:
1. Linearity : The relationship between predictors and the target is linear.
2. Independence : Observations are independent of each other
3. Homoscedacity : Constant variance of features.
4. Normality of features : Variables are normally distributed.

How it works
- Cost function - linear regression works by minimizing the cost function which measures the error between the predicted values and the actual vales,which essentially finds the line that best fits the data.
- Gradient descent optimization : is a way to minimize the cost function. It is an interactive optimization algorithm used to update the model's parameters(the y intercept and slopes)to minimize the error.At each interaction, the algorithms adjust the co-efficients to gradually minimize the error.

EVALUATION METRICES
1. Mean Squared Error : meaures the average squared difference between the actual and the predicted values
2. Mean Absolute Error : measures the average difference between the actual and predicted values.(IGNORE THE NEGATIVES)
3. r-squared :meaures the proportion of variance in tyhe dependent variable that is explained by the independent variables.

Applications:
- Predicting house prices based on features like square footage, no.of rooms,location etc
- Estimating sales revenue based on advertising spend across different channels.
- Modelling the relationship between a company's profit and factors like marketing,expenditure etc


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error

data = pd.DataFrame({
    "SquareFeet" : [1500,1600,1800,1900,2100,2200,2300,2500,2600,2800],
    "Price" : [3000, 3200,3400,3600,3800,4000,4200,4400,4600, 5000]
})

x = data[["SquareFeet"]]
y = data["Price"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

model = LinearRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print("Mean Squared Error:",mse)
print("Mean Absolute Error:",mae)
print("R-squared:",r2)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_train,y_train,color="blue",label="Training Data")
plt.scatter(x_test,y_test,color="red",label="Testing Data")

plt.plot(x_train, model.predict(x_train), color="green", label="Linear Regression")
plt.xlabel("Square Feet")
plt.ylabel("Price")
plt.title("Linear Regression Model")
plt.legend()
plt.show()


```
